## Conceptual Understanding 

### Q1. What are Supervised Learning Algorithms?

### Supervised Learning Algorithms are machine learning algorithms that learn from labeled data, where both input features and correct output values are provided. The model learns the relationship between inputs and outputs to make predictions on new data. Examples include Linear Regression, Logistic Regression, Decision Trees, and Support Vector Machines.

### Q2. Difference between Regression Algorithms and Classification Algorithms.

### Regression algorithms predict continuous numerical values such as house prices, sales, or temperatures. Classification algorithms predict categorical outcomes such as Yes/No, Spam/Not Spam, or Disease/No Disease. Linear Regression is an example of regression, while Logistic Regression is an example of classification.

### Q3. Explain Simple Linear Regression.

### Simple Linear Regression is a supervised learning algorithm that models the relationship between one independent variable and one dependent variable. It fits a straight line to the data using the equation y=mx+b, where m is the slope and b is the intercept. It is commonly used for prediction. 

### Q4. List and Explain the Assumptions of Linear Regression

### A. Linearity: There should be a linear relationship between input and output variables.
### B. Independence: Observations should be independent of each other.
### C. Homoscedasticity: The variance of errors should remain constant.
### D. Normality: Residual errors should be approximately normally distributed.
### E. No Multicollinearity: Independent variables should not be highly correlated with each other.

### Q5. What is the Bias-Variance Tradeoff?

### The bias-variance tradeoff refers to balancing model simplicity and complexity. High bias models are too simple and may underfit the data, while high variance models are too complex and may overfit. The goal is to find a model that minimizes both bias and variance to achieve good prediction performance on unseen data.

### Q6. Explain Overfitting and Underfitting with Examples.

### Underfitting occurs when a model is too simple to capture patterns in the data. For example, A regression model estimates home values relying strictly on square foot. Because it lacks other essential features like location, neighborhood safety scores, or building age, it yields heavily inaccurate estimates for almost every home.

### Overfitting occurs when a model learns the training data too closely, including outliers. For example, a very high-degree Polynomial Regression model may perform exceptionally well on training data but poorly on new test data. The ideal model requires a balance between underfitting and overfitting.
